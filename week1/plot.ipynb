```python
import matplotlib.pyplot as plt
import os
```

# Function that plots

```python
def plot_fn(algo, size, fn_calls, time):
    plt.figure(0)
    ax = plt.gca()
    ax.ticklabel_format(useOffset=False, style='plain')
    ax.get_xaxis().get_major_formatter().set_scientific(False)
    plt.plot(size, fn_calls, label='Function Calls')
    plt.xlabel('Size')
    plt.ylabel('Number of Function Calls')
    plt.title('Number of Function Calls vs Size for ' + algo)
    plt.legend()
    plt.savefig(os.path.join('plots', f'{algo}-fn_calls.png'))
    plt.show()
    plt.figure(1)
    ax = plt.gca()
    ax.ticklabel_format(useOffset=False, style='plain')
    ax.get_xaxis().get_major_formatter().set_scientific(False)
    plt.plot(size, time, label='Time')
    plt.xlabel('Size')
    plt.ylabel('Time (in seconds)')
    plt.title('Time vs Size for ' + algo)
    plt.legend()
    plt.savefig(os.path.join('plots', f'{algo}-time.png'))
    plt.show()
```

# Bubble Sort

```python
bubble_time = [6.594, 674.582, 14118.869]
bubble_fn_calls = [91944, 836565, 8054481]
plot_fn('Bubble Sort', [10000, 100000, 1000000], bubble_fn_calls, bubble_time)
```

# Quick Sort

```python
quick_time = [0.128, 1.537, 17.452, 229.162]
quick_fn_calls = [259645, 3014250, 33725000, 411841970]
plot_fn('Quick Sort', [10000, 100000, 1000000, 10000000], quick_fn_calls, quick_time)
```

# Counting Sort

```python
counting_time = [0.015, 0.025, 0.126, 1.135]
counting_fn_calls = [5678, 5324, 5059, 5694]
plot_fn('Counting Sort', [10000, 100000, 1000000, 10000000], counting_fn_calls, counting_time)
```

# Merge Sort

```python
merge_time = [0.144, 1.481, 16.046, 178.735]
merge_calls = [40741, 400741, 4000741, 40000741]
plot_fn('Merge Sort', [10000, 100000, 1000000, 10000000], merge_calls, merge_time)
```
